### This notebook loads and plots the agent's evaluation scores during training

```python
import pickle
import time
import random

import matplotlib.pyplot as plt
import numpy as np
```

```python
alien1 = pickle.load(open("baseline-QR-DQN/results/alien_scores",'rb'))
amidar1 = pickle.load(open("baseline-QR-DQN/results//amidar_scores",'rb'))
assault1 = pickle.load(open("baseline-QR-DQN/results/assault_scores",'rb'))
asterix1 = pickle.load(open("baseline-QR-DQN/results/asterix_scores",'rb'))
breakout1 = pickle.load(open("baseline-QR-DQN/results/breakout_scores",'rb'))

alien2 = pickle.load(open("thompson-QR-DQN/results/alien_scores",'rb'))
amidar2 = pickle.load(open("thompson-QR-DQN/results/amidar_scores",'rb'))
assault2 = pickle.load(open("thompson-QR-DQN/results/assault_scores",'rb'))
asterix2 = pickle.load(open("thompson-QR-DQN/results/asterix_scores",'rb'))
breakout2 = pickle.load(open("thompson-QR-DQN/results/breakout_scores",'rb'))

steps = np.arange(50) * 1000000
```

```python
plt.figure(figsize=(5,4))

plt.title("Alien", size=20)
plt.xlabel("Frames", size=20)
plt.ylabel("Score", size=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.plot(steps,alien1)
plt.plot(steps,alien2)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,4))

plt.title("Amidar", size=20)
plt.xlabel("Frames", size=20)
plt.ylabel("Score", size=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.plot(steps,amidar1)
plt.plot(steps,amidar2)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,4))

plt.title("Assault", size=20)
plt.xlabel("Frames", size=20)
plt.ylabel("Score", size=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.plot(steps,assault1)
plt.plot(steps,assault2)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,4))

plt.title("Asterix", size=20)
plt.xlabel("Frames", size=20)
plt.ylabel("Score", size=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.plot(steps,asterix1)
plt.plot(steps,asterix2)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,4))

plt.title("Breakout", size=20)
plt.xlabel("Frames", size=20)
plt.ylabel("Score", size=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.plot(steps,breakout1)
plt.plot(steps,breakout2)

plt.tight_layout()
plt.show()
```
